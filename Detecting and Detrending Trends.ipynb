# Task 1: Detecting trends using Hodrick-Prescott Filter:

```
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```
df = pd.read_csv('/content/shampoo.csv', index_col=0, parse_dates=True)
```

```
Sales_cycle, Sales_trend = hpfilter(df['Sales'], lamb=1600)

plt.figure(figsize=(15, 6))
Sales_trend.plot()
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Shampoo Sales Trend')
plt.show()
```

Note: The code utilizes Pandas, Matplotlib, and Statsmodels libraries to analyze monthly shampoo sales data. It loads the data from 'shampoo.csv' into a Pandas DataFrame and applies the Hodrick-Prescott (HP) filter to separate the sales data into its cyclical and trend components. The resulting trend component is then visualized using Matplotlib, displaying the sales trend over time. The dataset includes 'Month' and 'Sales' columns, with 'Month' representing dates in a '01-Month' format and 'Sales' containing corresponding sales values for each month

# Task 2-A: Detrending a Time Series (Using Pandas):

```
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('/content/shampoo.csv', index_col=0, parse_dates=True)

X = df['Sales']
diff = list()

for i in range(1, len(X)):
  value = X[i] - X[i-1]
  diff.append(value)

plt.figure(figsize=(15, 6))
plt.plot(diff)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Detrending Shampoo Sales using Pandas Differencing')
plt.show()
```

Note: The code utilizes Pandas and Matplotlib libraries to analyze monthly shampoo sales data. It reads the data from 'shampoo.csv' into a Pandas DataFrame and calculates the difference between consecutive sales values to detrend the data. The resulting differences are then plotted using Matplotlib, visualizing the detrended sales pattern over time. The dataset contains 'Month' and 'Sales' columns, with 'Month' representing dates and 'Sales' containing corresponding sales values for each month.

```
import pandas as pd
import matplotlib.pyplot as plt

series = pd.read_excel('/content/India_Exchange_Rate_Dataset.xls', index_col=0, parse_dates=True)
X = series.values
diff = list()

for i in range(1, len(X)):
    value = X[i] - X[i-1]
    diff.append(value)

plt.figure(figsize=(15, 6))
plt.plot(diff)
plt.title('Detrending Exchange Rates using Pandas Differencing', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS exchange rate')
plt.show()
```

# Task 2-B: Detrending a Time Series (Using .diff()):

```
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

def Parser(x):
  return datetime.strptime('190' + x, '%Y-%m')

df = pd.read_csv('/content/shampoo.csv', header=0, parse_dates=[0], index_col=0, date_parser=Parser)
diff = df.Sales.diff()

plt.figure(figsize=(15, 6))
plt.plot(diff)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Detrending Shampoo Sales using .diff()')
plt.show()
```

Note: This code snippet utilizes Pandas and Matplotlib libraries to analyze monthly shampoo sales data. It reads the data from 'shampoo.csv' into a Pandas DataFrame, parsing dates with a custom function. The code then calculates the difference between consecutive sales values using the .diff() method, effectively detrending the data. Finally, it generates a plot with Matplotlib, visualizing the detrended shampoo sales pattern over time. The dataset includes 'Month' and 'Sales' columns, with 'Month' representing dates, and 'Sales' containing corresponding sales values for each month.

# Task 2-C: Detrending a Time Series (Using SciPy):

```
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('/content/shampoo.csv', index_col=0, parse_dates=True)
detrended = signal.detrend(df['Sales'].values)

plt.figure(figsize=(15, 6))
plt.plot(detrended)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Detrending Shampoo Sales using Scipy Signal', fontsize=16)
plt.show()
```

# Task 2-D: Detrending a Time Series (Using HP Filter):

```
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.filters.hp_filter import hpfilter
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('/content/shampoo.csv', index_col=0, parse_dates=True)
Sales_cycle, Sales_trend = hpfilter(df['Sales'], lamb=1600)
df['trend'] = Sales_trend
detrended = df['Sales'] - df['trend']

plt.figure(figsize=(15, 6))
plt.plot(detrended)
plt.title('Detrending Shampoo Sales using HP Filter', fontsize=16)
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.show()
```

Note: This code snippet employs Pandas, Matplotlib, and Statsmodels libraries to analyze monthly shampoo sales data. It loads the data from 'shampoo.csv' into a Pandas DataFrame and applies the Hodrick-Prescott (HP) filter to separate the sales data into its cyclical and trend components. The code then calculates the detrended sales by subtracting the trend component. It generates a plot with Matplotlib, showcasing the detrended shampoo sales pattern over time. The dataset contains 'Month' and 'Sales' columns, with 'Month' representing dates and 'Sales' containing corresponding sales values for each month. The x-axis displays the 'Month' values with a 45-degree rotation for better readability.

```
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.filters.hp_filter import hpfilter
import warnings

warnings.filterwarnings("ignore")

df = pd.read_excel(r'India_Exchange_Rate_Dataset.xls', index_col=0, parse_dates=True)
EXINUS_cycle, EXINUS_trend = hpfilter(df['EXINUS'], lamb=1600)
df['trend'] = EXINUS_trend
detrended = df['EXINUS'] - df['trend']

plt.figure(figsize=(15, 6))
plt.plot(detrended)
plt.title('Detrending Exchange Rates using HP Filter', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS')
plt.show()
```
